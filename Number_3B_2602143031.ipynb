```python
import numpy as np
import soundfile as sf
import IPython.display as ipd
from nemo.collections.tts.models import FastPitchModel, HifiGanModel
print("Imported successfully!")
```

```python
# Load FastPitch model
print("Loading FastPitch model...")
fastpitch = FastPitchModel.from_pretrained(model_name="tts_en_fastpitch").cuda()
print("FastPitch loaded!")

# Load HiFi-GAN vocoder
print("Loading HiFi-GAN vocoder...")
vocoder = HifiGanModel.from_pretrained(model_name="tts_en_hifigan").cuda()
print("HiFi-GAN loaded!")

print("\n" + "="*40)
print("MODELS READY!")
print("="*40)
```

```python
# Input text
input_text = "Text to Speech or TTS plays an important role in helping the economic sector to remind customers personally in an organized and effective manner."

print("Input text:")
print(f"   {input_text}")
print("\nGenerating speech")

# Parse text and generate spectrogram
parsed = fastpitch.parse(input_text)
spectrogram = fastpitch.generate_spectrogram(tokens=parsed)

# Convert spectrogram to audio
audio = vocoder.convert_spectrogram_to_audio(spec=spectrogram)
audio_numpy = audio.to('cpu').detach().numpy().flatten()

print("Speech generated!")
```

```python
output_filename = "eng_tts_output.wav"
sf.write(output_filename, audio_numpy, 22050)

print(f" Audio saved as: {output_filename}")
print(f"   Sample rate: 22050 Hz")
print(f"   Duration: {len(audio_numpy) / 22050:.2f} seconds")
```

```python
print("Playing TTS output:")
ipd.Audio(output_filename)
```
